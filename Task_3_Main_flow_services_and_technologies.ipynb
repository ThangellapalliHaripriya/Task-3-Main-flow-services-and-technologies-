Create a bar chart and a line chart using Matplotlib to visualize data from a Pandas DataFrame. Customize the charts with labels, titles, and legends.

```python
#importing pandas library to read the data from the csv file
import pandas as pd
#importing matplotlib to visualize the data
import matplotlib.pyplot as plt
```

```python
#loading  the data set
df = pd.read_csv('/content/householdtask3.csv.crdownload')
```

```python
df.head(5)
```

```python
#bar chart visuals
plt.figure(figsize=(10, 6))
plt.bar(df['year'] - 0.2, df['own'], width=0.4, label='Own')
plt.bar(df['year'] + 0.2, df['own_wm'], width=0.4, label='Own WM')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of Households')
plt.title('Household Ownership and Washing Machine Ownership Over Time')
plt.xticks(df['year'])

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
#line chart visuals
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['income'], marker='d', label='Income', color='blue')
plt.plot(df['year'], df['expenditure'], marker='d', label='Expenditure', color='red')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Household Income and Expenditure Over Time')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
